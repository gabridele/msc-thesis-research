```python
# questo funziona. i extracted arrays containing volume for each roi

# Step 2: Get the unique ROI labels
roi_labels = np.unique(atlas)
roi_labels = roi_labels[roi_labels != 0]  # Assuming 0 is not a valid ROI

# Step 3: Initialize a dictionary to store coordinates for each ROI
roi_coordinates = {}

# Step 4: Extract coordinates for each ROI
for roi in roi_labels:
    # Create a mask for the current ROI
    roi_mask = atlas == roi
    
    # Get the coordinates of the voxels in this ROI
    coordinates = np.column_stack(np.where(roi_mask))
    
    # Store the coordinates in the dictionary
    roi_coordinates[roi] = coordinates

# Print the number of ROIs and the coordinates for an example ROI
print("Number of ROIs:", len(roi_coordinates))
example_roi = roi_labels[59]
print(f"Coordinates of ROI {example_roi}:\n", roi_coordinates[example_roi])
```
