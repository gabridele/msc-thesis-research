```python
import numpy as np
!pwd
```

```python
matrix10k = np.loadtxt("../spreading_dynamics_clinical/derivatives/sub-60033/dwi/association_matrix_sub-60033.csv", delimiter=',')
matrix100k = np.loadtxt("../spreading_dynamics_clinical/derivatives/sub-60033/dwi/association_matrix_sub-60033100k.csv", delimiter=',')
```

```python
correlations = [np.corrcoef(matrix10k[i], matrix100k[i])[0, 1] for i in range(matrix10k.shape[0])]
```

```python
for i, corr in enumerate(correlations):
    print(f"Correlation for column {i}: {corr}")
```

```python
mean_correlation = np.mean(correlations)
print(mean_correlation)
```
