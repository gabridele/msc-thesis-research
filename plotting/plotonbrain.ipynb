```python
import numpy as np
import glob
import nibabel as nib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

```python
dir = "/home/gabridele/Desktop/irbio_folder/spreading_dynamics_clinical/HC_emp/"

#list all files with glob
files = glob.glob(dir + 'emp*.npy')

#load files and stack them into single array
arrays = [zscore(np.load(file)) for file in files]
print(arrays)
mean_array = np.mean(arrays, axis=0)

#save mean array to new file
output = "avgd_emp_hc.npy"
np.save(dir + output, mean_array)
```

```python
#load atlas data
atlas_path = "/home/gabridele/Desktop/irbio_folder/spreading_dynamics_clinical/derivatives/templates/Schaefer2018_400Parcels_Tian_Subcortex_S4_1mm_2009c_NLinAsymm.nii.gz"
atlas_img = nib.load(atlas_path)
atlas_data = atlas_img.get_fdata()
```

```python
#zscored_vols = vols.apply(zscore, axis=1)
```

```python
stats_data = np.zeros_like(atlas_data)
for roi_index in range(1, len(mean_array) + 1):
    roi_indices = np.where(atlas_data == roi_index)
    stats_data[roi_indices] = mean_array[roi_index - 1]

stats_img = nib.Nifti1Image(stats_data, atlas_img.affine, header=atlas_img.header)
nib.save(stats_img, 'emp_hc_values_mapped.nii.gz')
```
